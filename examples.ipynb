# Stocks Portfolio Examples:

```python
# imports
from portfolio import Portfolio
```

```python
# Create a portfolio and add one stock and one ETF into it
p = Portfolio()
p.buy("fb", "stock", price=404.5, amount=5)
p.buy("VTI", "etf", price=240, amount=5)
```

```python
# it's also possible to sell partial amount of the stock
p.sell("fb", amount=1)
```

```python
# show the current portfolio values
p.details()
```

```python
# get trend of the portfolio performance in the last month
p.plot_hist()
```

```python
# see the details for the portfolio performance in the last month:
p.get_hist_value()
```

```python
# compare to the S&P 500 index, but this can be also compared to a stock or an etf
p.plot_compare_to(ticker="S&P 500", p_type="index")
```

```python
# perform a basic analysis of the portfolio with the last day values of gain, change % in the last day, etc
p.basic_analysis()
```

```python
# Get technical recommendations from investing and Yahoo Finance
p.gen_recommendations()
```
